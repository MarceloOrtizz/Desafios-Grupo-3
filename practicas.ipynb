```python
import pandas as pd

# DataFrame con datos duplicados
data = {
    'Nombre': ['Ana', 'Juan', 'Ana', 'Pedro', 'María', 'Juan'],
    'Edad': [25, 30, 25, 40, 28, 30],
    'Ciudad': ['Madrid', 'Barcelona', 'Madrid', 'Valencia', 'Sevilla', 'Barcelona']
}

df = pd.DataFrame(data)
df

```

```python
# Eliminar filas duplicadas basadas en todas las columnas
df_sin_duplicados = df.drop_duplicates()
df_sin_duplicados
```

```python
# Eliminar filas duplicadas basadas solo en la columna 'Nombre'
df_sin_duplicados_nombre = df.drop_duplicates(subset=['Nombre'])
df_sin_duplicados_nombre
```

```python
# Eliminar filas duplicadas manteniendo la última aparición
df_sin_duplicados_last = df.drop_duplicates(subset=['Nombre'], keep='last')
df_sin_duplicados_last
```

```python
# Modificar el DataFrame original eliminando las filas duplicadas
df.drop_duplicates(inplace=True)
```
